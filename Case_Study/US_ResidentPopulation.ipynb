Dataset: "Annual Estimates of the Resident Population by Single Year of Age
and Sex for the United States."

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/flaviovdf/evcomp2018/master/data/nc-est2015-agesex-res.csv")
df.head(10)
```

```python
print(f"The size is {df.shape}")
```

```python
print(f"The Description :\n")
df.describe()
```

```python
df = df.rename(columns={"POPESTIMATE2010":"2010","POPESTIMATE2014":"2014"})
```

```python
print("The first 10 rows")
df.head(10)
```

```python
df.iloc[[14,21]]
```

```python
df["change"] = df["2014"] - df["2010"]
df["percentage_change"] = (df["change"]/df["2010"])*100
```

```python
df.head(5)
```

```python
graying_population = df[(df.AGE >= 67) & (df.AGE <= 71)]
                                         
print("Graying of US population:\n", graying_population)
```

```python
infants_data = df[df.AGE == 0]
male_infants = infants_data['SEX'][infants_data["SEX"] == 1].count()
female_infants = infants_data['SEX'][infants_data["SEX"] == 2].count()
infants_gender_ratio = female_infants / male_infants

print("Infants gender ratio:", infants_gender_ratio)
```

```python
infants_data['SEX'][infants_data["SEX"] == 1]
```

```python
female_to_male_ratio_2_to_4 = [(df.loc[df["SEX"] == 0].count()/ df.loc[df["SEX"] == 1].count()).between(2, 4)]
print("Years with female to male ratio between 2 and 4:\n", female_to_male_ratio_2_to_4)
```

```python
df.loc[df["SEX"] == '1']
```

```python
plt.plot(df.AGE, df['SEX'][df["SEX"] == 1] / df['SEX'][df["SEX"] == 2], marker='o', linestyle='-', color='b', label='Female-to-Male Ratio')
plt.xlabel('Age')
plt.ylabel('Female-to-Male Ratio')
plt.title('Gender Ratio versus Age')
```
